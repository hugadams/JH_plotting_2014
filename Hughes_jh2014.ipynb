### Abstract

Nanotechnology and image processing are two of the most rapidly growing, cross-disciplinary research fields.  For example, gold nanoparticles alone are found in novel applications from photovoltaics[1], to protein sensing[2], and even to boiling water[3].  Fascinating progress in areas like computer vision ensure that image processing will continue to attract both eruidite and commercial interest in the foreseeable future.

High-resolution microscopy resides at the intersection of nanotech. and image processing.  This is not surprising, since the morphological characteristics of nanomaterials, at the individual particle and ensemble scales, completely determines the properties of the composite.  One would expect to find a considerable body of knowledge and software geared towards nanosystem image processing; however, this is not the case.  By contrast, consider the knowledgebase and software infrastcture of cell profiling[4] which vastly overshadows its nanoimaging counterpart, despite almost identical considerations in workflow and constructs.  Using scientific Python, especially scikit-image, we have begun addressing the dearth.

We sought ought to create a guide for nanomaterial image processing, but didn't want to make subjective assessments based on our sample images.  Using scikit-image and a complementary particle analysis library, pyparty, we set out to build artificial electron microscope images.  These would allow for quantitive assesments of preprocessing, segmentation and object-identification algorithms[5], motivated in the context of nanoparticle domain.  The images turned out better than expected, featuring multi-species particles of varying brightness and orientation patterend over a shadowed background.  Realistic particle edges were obtained with guassian smoothing, accompanied by normal noise via numpy.  

Shown below is the final **(3072x2304)** image, and its brightness distribution, decomposed into the various components (noise, background, particles etc...).  This plot conveys how each component of the image contributes to the grayscale histogram.  Shown individually are color-labeld particle species, smootehd edges, noise (false color to show off intensity distribution) and background sans noise and particles.
